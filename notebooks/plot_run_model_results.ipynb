In this notebook we present the results of the pre-existing benchmark model training for the LST CNN-RNN configuration.

### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Open the files and plot  the figures

The input of the functions are the .csv files exported in Tensorboard.

```python
def plot_run_eval(file):
    df = pd.read_csv(file)
    plt.figure(figsize=(10,5))
    plt.scatter(df['Step'], df['Value'], marker = 'o', s=10, label='LST')
    plt.plot(df['Step'], df['Value'], alpha=0.5, label=_)
    for i,j in zip(df['Step'], df['Value']):
        plt.text(i, j, (str(j)[0:5]), fontsize=8, alpha = 0.8)
    plt.legend(loc = 0)
    plt.title(file[:-4])    
    plt.show
    
    plt.savefig(file[:-4] + '.pdf')
    return None
```

```python
def plot_run(file):
    df = pd.read_csv(file)
    
    plt.figure(figsize=(10,5))
    plt.scatter(df['Step'], df['Value'], marker = 'o', s=0.5, c='orange', label='LST')
    plt.plot(df['Step'], df['Value'], alpha=0.5, c='orange', label=_)
    plt.legend(loc = 0)

    plt.title(file[:-4])    
    plt.show
    
    plt.savefig(file[:-4] + '.pdf')
    return None
```

### Results

```python
plot_run_eval('run_logs_run00_eval_validation-tag-accuracy.csv')
```

```python
plot_run_eval('run_logs_run00_eval_validation-tag-auc.csv')
```

```python
plot_run_eval('run_logs_run00_eval_validation-tag-loss.csv')
```
